```python
##Generate lists for S/N and Count Rates
ASNIDs = []
average_count_rates = []
signal_to_noise = []
```

```python
##Convert ASN_IDs to lowercase
for IDnum in ASN_IDs:
    ID_lower = IDnum.lower()
    ASNIDs.append(ID_lower)
```

```python
for ID_num in ASNIDs:
    if ID_num == 'none' or ID_num == 'lbx502030':
        signal_to_noise.append('N/A')
        average_count_rates.append('N/A')
    else:
        filepath = '/grp/hst/cos2/cosmo/' + str(rootname_PID_dict[list(ROOTNAME_ASN_dict.keys())[list(ROOTNAME_ASN_dict.values()).index(ID_num)]]) + '/' + ID_num + '_x1dsum.fits.gz'
        x1dsum_data = fits.getdata(filepath)
        x1dsum_detector = fits.getval(filepath, 'detector')
        for segment in x1dsum_data['segment']:
            # get the GROSS and WAVELENGTH arrays from the x1dsum
            grosscr = x1dsum_data['GROSS'][np.where(x1dsum_data['SEGMENT'] == segment)][0]
            wl = x1dsum_data['WAVELENGTH'][np.where(x1dsum_data['SEGMENT'] == segment)][0]
            # select out the wavelength range of Lyman Alpha (1215Angstroms)
            sel_la = np.where((wl < 1213) | (wl > 1217))
            # select only gross counts not in the Ly-a region
        gross_wo_lya = grosscr[sel_la]
        # find the average count rate only where the gross counts > 0
        avecr = np.mean(gross_wo_lya[np.where(gross_wo_lya > 0)])
        # find the S/N only where the gross counts > 0
        # multiply by sqrt(6) for a resel
        sn = avecr / np.std(gross_wo_lya[np.where(gross_wo_lya > 0)]) * np.sqrt(6)
        average_count_rates.append(avecr)
        signal_to_noise.append(sn)
```
